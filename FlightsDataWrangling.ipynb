```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("flight-delays/flights.csv")
```

```python
df.info()
```

```python
pd.set_option('display.max_columns', 84)
```

```python
df.head(10)
```

```python
df['CANCELLATION_REASON'].value_counts(dropna= False)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy= 'most_frequent')
df['CANCELLATION_REASON'] = imputer.fit_transform(df[['CANCELLATION_REASON']]).ravel()
df.head()
```

```python
df['AIR_SYSTEM_DELAY'].value_counts(dropna= False)
```

```python
df['AIR_SYSTEM_DELAY'].describe().round(3)
```

```python
imputer = SimpleImputer(missing_values=np.nan, strategy= 'mean')
df[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']]= imputer.fit_transform(df[['AIR_SYSTEM_DELAY', 'SECURITY_DELAY', 'AIRLINE_DELAY', 'LATE_AIRCRAFT_DELAY', 'WEATHER_DELAY']])
df.head()
```

```python
df.to_csv('flights.csv', index=False)
```
